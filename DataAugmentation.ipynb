### Interpolate another 2 frames between original 2 consecutive frames

```python
def getXinter(x, number): #create linearly spaced intervals between 2 consecutive coordinates
    lst=[]
    for i, _ in enumerate(x):
        try:
            value=x[i:i+2]
            temp=(np.linspace(value[0],value[1],num=number+2))
            lst.append(temp[:-1])
            #lst.append((value[0]+value[1])/2)
        except:
            temp=(np.linspace(x[i],0,num=number+2))
            lst.append(temp[:-1])
    return lst

augmented_rowData1=[]
augmented_rowData2=[]
augmented_rowData3=[]

for row in df.itertuples(index=False):
    
    
    framelist1=[]
    framelist2=[]
    framelist3=[]
    #print(row)
    for eachFrame in row:
        eachFrame=ast.literal_eval(eachFrame)
        x_coord=eachFrame[0::2]
        y_coord=eachFrame[1::2]
#         print(x_coord[0])
#         print(len(x_coord),len(y_coord))
        
        interp_func=interp1d(x_coord,y_coord,bounds_error=False, fill_value=0)
        
        x_inter=getXinter(x_coord,2) #2 represents adding 2 new coordinates in between existing.
        y_inter=interp_func(x_inter)
        
        flatten_x_inter= [eachCoord for coordList in x_inter for eachCoord in coordList]
        
        flatten_y_inter= [eachCoord for coordList in y_inter for eachCoord in coordList]
        flatten_y_inter=[0 if y!=y else y for y in flatten_y_inter]
        
        augmented_perFrame=list(zip(flatten_x_inter, flatten_y_inter))
        
        k1=[i for each in augmented_perFrame[0::3] for i in each]
        k2=[i for each in augmented_perFrame[1::3] for i in each]
        k3=[i for each in augmented_perFrame[2::3] for i in each]
        
        framelist1.append(k1)
        framelist2.append(k2) 
        framelist3.append(k3)
    augmented_rowData1.append(framelist1)
    augmented_rowData2.append(framelist2)
    augmented_rowData3.append(framelist3)
        
```
