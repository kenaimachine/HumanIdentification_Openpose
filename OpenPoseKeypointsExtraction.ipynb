## Run OpenPose

1.   Convert every video file from avi format to mp4, using convert_avi_to_mp4 function. Save mp4 video.

4.   Run Openpose
5.   Save pose overlayed video in folder 'Openpose'
6.   Save Keypoints in folder 'Keypoints'
